```python
# Dropping columns with a high percentage of missing values that are not critical for the analysis.
# The columns to be dropped are identified based on their relevance and the amount of missing data.
# Dropping columns with a high percentage of missing values that are not critical for the analysis.
# Only drop columns that exist in the DataFrame to avoid errors.
columns_to_drop = ['Flight_Number_Reporting_Airline', 'Tail_Number', 'Origin_Airport_ID', 
                   'Origin_Airport_SEQ_ID', 'Origin_City_Market_ID', 'Origin_State_Fips', 
                   'Origin_Wac', 'Dest_Airport_ID', 'Dest_Airport_SEQ_ID', 
                   'Dest_City_Market_ID', 'Dest_State_Fips', 'Dest_Wac']
df.drop(columns=[col for col in columns_to_drop if col in df.columns], inplace=True)
# Checking the DataFrame after dropping the columns
df.info()
```

```python
# Impute numerical columns with the median
numerical_cols_to_impute_median = [
    'Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries',
    'Total.Uninjured', 'Number.of.Engines'
]
for col in numerical_cols_to_impute_median:
    if col in df.columns:
        df[col].fillna(df[col].median(), inplace=True)

# Impute categorical columns with the mode
categorical_cols_to_impute_mode = [
    'Injury.Severity', 'Aircraft.damage', 'Amateur.Built', 'Engine.Type',
    'Purpose.of.flight', 'Weather.Condition', 'Make', 'Model', 'Registration.Number'
]
for col in categorical_cols_to_impute_mode:
    if col in df.columns:
        df[col].fillna(df[col].mode()[0], inplace=True)

# Drop rows with missing values for 'Event.Date', 'Location', and 'Country'
df.dropna(subset=[col for col in ['Event.Date', 'Location', 'Country'] if col in df.columns], inplace=True)

# Display the number of remaining missing values to confirm imputation and dropping
display(df.isnull().sum())
```

```python
# 1. Convert the 'Event.Date' column to datetime objects.
df['Event.Date'] = pd.to_datetime(df['Event.Date'])

# 2. Extract the year from the 'Event.Date' and store it in a new column named 'Event.Year'.
df['Event.Year'] = df['Event.Date'].dt.year

# 3. Clean and standardize the 'Make' and 'Model' columns by converting them to lowercase and stripping any leading/trailing whitespace.
df['Make'] = df['Make'].str.lower().str.strip()
df['Model'] = df['Model'].str.lower().str.strip()

# 4. Clean and extract numerical information from the 'Injury.Severity' column to represent the total number of injuries or fatalities.
# This column contains strings like 'Fatal(2)', 'Serious(1)', 'Minor(3)', 'Non-Fatal'.
# We can extract the number in parentheses if it exists and create a new column for total injuries.
import re
def extract_injury_count(severity_string):
    if isinstance(severity_string, str):
        match = re.search(r'\((\d+)\)', severity_string)
        if match:
            return int(match.group(1))
    return 0 # Return 0 for 'Non-Fatal', NaN, or other non-matching strings

df['Total_Extracted_Injuries'] = df['Injury.Severity'].apply(extract_injury_count)

display(df[['Event.Date', 'Event.Year', 'Make', 'Model', 'Injury.Severity', 'Total_Extracted_Injuries']].head())
display(df.info())
```

```python
# 1. Analyze the distribution of Injury.Severity
if 'Injury.Severity' in df.columns:
	plt.figure(figsize=(10, 6))
	sns.countplot(data=df, y='Injury.Severity', order=df['Injury.Severity'].value_counts().index)
	plt.title('Distribution of Injury Severity')
	plt.xlabel('Number of Accidents')
	plt.ylabel('Injury Severity')
	plt.show()

# 2. Examine the relationship between Aircraft.damage and Injury.Severity
if 'Aircraft.damage' in df.columns and 'Injury.Severity' in df.columns:
	plt.figure(figsize=(12, 8))
	sns.countplot(data=df, x='Aircraft.damage', hue='Injury.Severity', palette='viridis')
	plt.title('Relationship between Aircraft Damage and Injury Severity')
	plt.xlabel('Aircraft Damage')
	plt.ylabel('Number of Accidents')
	plt.xticks(rotation=45, ha='right')
	plt.legend(title='Injury Severity')
	plt.tight_layout()
	plt.show()

# 3. Investigate the distribution of accidents across different Weather.Condition categories
if 'Weather.Condition' in df.columns:
	plt.figure(figsize=(10, 6))
	sns.countplot(data=df, y='Weather.Condition', order=df['Weather.Condition'].value_counts().index, palette='plasma')
	plt.title('Distribution of Accidents by Weather Condition')
	plt.xlabel('Number of Accidents')
	plt.ylabel('Weather Condition')
	plt.show()

# 4. Explore the trends of accidents over the years
if 'Event.Year' in df.columns:
	accidents_by_year = df['Event.Year'].value_counts().sort_index()
	plt.figure(figsize=(12, 6))
	sns.lineplot(x=accidents_by_year.index.astype(int), y=accidents_by_year.values)
	plt.title('Trend of Accidents Over Years')
	plt.xlabel('Year')
	plt.ylabel('Number of Accidents')
	plt.grid(True)
	plt.show()
```

```python
# Group by 'Make' and 'Model' and calculate aggregated metrics robustly

agg_dict = {
    'total_accidents': ('Event.Date', 'count')
}

if 'Total.Fatal.Injuries' in df.columns:
    agg_dict['total_fatal_injuries'] = ('Total.Fatal.Injuries', 'sum')
if 'Total.Serious.Injuries' in df.columns:
    agg_dict['total_serious_injuries'] = ('Total.Serious.Injuries', 'sum')
if 'Total.Minor.Injuries' in df.columns:
    agg_dict['total_minor_injuries'] = ('Total.Minor.Injuries', 'sum')
if 'Total.Uninjured' in df.columns:
    agg_dict['total_uninjured'] = ('Total.Uninjured', 'sum')
if 'Number.of.Engines' in df.columns:
    agg_dict['average_engines'] = ('Number.of.Engines', 'mean')
if 'Aircraft.damage' in df.columns:
    agg_dict['destroyed_percentage'] = ('Aircraft.damage', lambda x: ((x == 'Destroyed').sum() / max(x.dropna().shape[0], 1)) * 100)
if 'Weather.Condition' in df.columns:
    agg_dict['imc_percentage'] = ('Weather.Condition', lambda x: ((x == 'IMC').sum() / max(x.dropna().shape[0], 1)) * 100)

risk_metrics = df.groupby(['Make', 'Model']).agg(**agg_dict).reset_index()

# Calculate additional risk metrics if columns exist
if all(col in risk_metrics.columns for col in ['total_fatal_injuries', 'total_serious_injuries', 'total_minor_injuries']):
    risk_metrics['total_injuries'] = risk_metrics['total_fatal_injuries'] + \
                                     risk_metrics['total_serious_injuries'] + \
                                     risk_metrics['total_minor_injuries']
    risk_metrics['injury_rate_per_accident'] = risk_metrics['total_injuries'] / risk_metrics['total_accidents']
    risk_metrics['fatal_injury_rate_per_accident'] = risk_metrics['total_fatal_injuries'] / risk_metrics['total_accidents']

# Display the head of the new DataFrame
display(risk_metrics.head())
```

```python
# 1. Top 20 Aircraft by Accident Count
top_20_accidents = risk_metrics.sort_values(by='total_accidents', ascending=False).head(20)
top_20_accidents['Aircraft'] = top_20_accidents['Make'] + ' ' + top_20_accidents['Model']

plt.figure(figsize=(10, 8))
sns.barplot(data=top_20_accidents, x='total_accidents', y='Aircraft', palette='viridis')
plt.title('Top 20 Aircraft by Accident Count')
plt.xlabel('Number of Accidents')
plt.ylabel('Aircraft Make and Model')
plt.tight_layout()
plt.show()
```

```python
# 2. Lowest Risk Aircraft by Injury Rate
lowest_risk_aircraft = risk_metrics.sort_values(by='injury_rate_per_accident', ascending=True)
lowest_risk_aircraft_top_20 = lowest_risk_aircraft.head(20).copy()
lowest_risk_aircraft_top_20['Aircraft'] = lowest_risk_aircraft_top_20['Make'] + ' ' + lowest_risk_aircraft_top_20['Model']

plt.figure(figsize=(10, 8))
sns.barplot(data=lowest_risk_aircraft_top_20, x='injury_rate_per_accident', y='Aircraft', palette='viridis')
plt.title('Top 20 Lowest Risk Aircraft by Injury Rate per Accident')
plt.xlabel('Injury Rate per Accident')
plt.ylabel('Aircraft Make and Model')
plt.tight_layout()
plt.show()
```

```python
# 3. Injury Severity Distribution for Top 5 Highest Accident Aircraft
top_5_accidents_aircraft = top_20_accidents.head(5)

fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(25, 5))
axes = axes.flatten()

for i, (index, row) in enumerate(top_5_accidents_aircraft.iterrows()):
    make = row['Make']
    model = row['Model']

    # Filter the original df for the specific aircraft
    aircraft_df = df[(df['Make'] == make) & (df['Model'] == model)]

    # Calculate the value counts of Injury.Severity
    severity_counts = aircraft_df['Injury.Severity'].value_counts()

    # Create a pie chart
    axes[i].pie(severity_counts, labels=severity_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2', len(severity_counts)))
    axes[i].set_title(f'{make.title()} {model.title()}')

plt.tight_layout()
plt.show()
```
